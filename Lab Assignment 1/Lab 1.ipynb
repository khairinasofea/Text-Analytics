```python
import re
import json
import requests
import csv
from datetime import datetime

# Extract shop ID and item ID from the URL
url = 'https://shopee.com.my/Sony-Bone-Conduction-Bluetooth-Earphone-Earring-Wireless-Ear-Clip-Earbuds-Sound-Earcuffs-Sport-Headset-With-Mic-i.48959264.20773737808?sp_atk=b253076a-d05e-47a0-8175-34a80703b02b&xptdk=b253076a-d05e-47a0-8175-34a80703b02b'
match = re.search(r'i\.(\d+)\.(\d+)', url)
if not match:
    print("Invalid URL")
    exit()


shop_id, item_id = match.group(1), match.group(2)

# Set the desired number of pages to scrape
desired_pages = 3

# Open CSV file for writing
with open('test1.csv', 'w', newline='', encoding='utf-8') as csvfile:
    fieldnames = ['Username', 'Date', 'Comment']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()

    # Loop through desired number of pages
    for page in range(desired_pages):
        # Fetch ratings data from Shopee API
        offset = page * 20  # Calculate offset based on page number
        ratings_url = f'https://shopee.com.my/api/v2/item/get_ratings?filter=0&flag=1&itemid={item_id}&limit=20&offset={offset}&shopid={shop_id}&type=0'
        data = requests.get(ratings_url).json()

        # Extract and write ratings to CSV
        for rating in data['data']['ratings']:
            author_username = rating['author_username']
            
            timestamp = rating['mtime']
            date = datetime.utcfromtimestamp(timestamp).strftime('%Y-%m-%d')  # Convert to desired format
            
            # Combine all lines of comment into one string
            comment_lines = [line.strip() for line in rating['comment'].split('\n')]
            comment = ' '.join(comment_lines)
            
            # Write to CSV file
            writer.writerow({'Username': author_username, 'Date': date, 'Comment': comment})

print("Scraping complete!")
```

```python
import re
import json
import requests
import csv
from datetime import datetime

# Extract shop ID and item ID from the URL
url = 'https://shopee.com.my/Sony-Bone-Conduction-Bluetooth-Earphone-Earring-Wireless-Ear-Clip-Earbuds-Sound-Earcuffs-Sport-Headset-With-Mic-i.48959264.20773737808?sp_atk=b253076a-d05e-47a0-8175-34a80703b02b&xptdk=b253076a-d05e-47a0-8175-34a80703b02b'
match = re.search(r'i\.(\d+)\.(\d+)', url)
if not match:
    print("Invalid URL")
    exit()

shop_id, item_id = match.group(1), match.group(2)

# Set the desired number of pages to scrape
desired_pages = 3

# Open CSV file for writing
with open('husnin.csv', 'w', newline='', encoding='utf-8') as csvfile:
    fieldnames = ['Username', 'Date', 'Comment']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()

    # Loop through desired number of pages
    for page in range(desired_pages):
        # Fetch ratings data from Shopee API
        offset = page * 6  # Calculate offset based on page number
        ratings_url = f'https://shopee.com.my/api/v2/item/get_ratings?filter=0&flag=1&itemid={item_id}&limit=20&offset={offset}&shopid={shop_id}&type=0'
        data = requests.get(ratings_url).json()

        # Extract and write ratings to CSV
        for rating in data['data']['ratings']:
            author_username = rating['author_username']
            
            timestamp = rating['mtime']
            date = datetime.utcfromtimestamp(timestamp).strftime('%Y-%m-%d')  # Convert to desired format
            
            # Combine all lines of comment into one string
            comment_lines = [line.strip() for line in rating['comment'].split('\n')]
            comment = ' '.join(comment_lines)
            
            # Write to CSV file
            writer.writerow({'Username': author_username, 'Date': date, 'Comment': comment})

        # If the current page is the last page we want to scrape, break out of the loop
        if page == desired_pages - 1:
            break

print("Scraping complete!")
```
